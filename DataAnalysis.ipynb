```python
gb= df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(gb,annot=True)
plt.title("Parents education is have the good impact on student's course")
plt.show()
```

```python
#from the above graph we have concluded that the parents education is have the good impact on student's course
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(gb1,annot=True)
plt.title("Relationship betweeen parent's marital status and student's score")
plt.show()
```

```python
#from the above chart we have analysed that the marital status will not impact on student's education
```
